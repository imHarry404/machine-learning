```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
```

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students
classes2_student=[40,5,20,20,10]
classes3_student=[35,5,30,15,15]
```

### plotting bar chart using these data, data can be in any format

```python
plt.bar(classes,classes1_student)
```

### horizontally

```python
plt.barh(classes,classes1_student)
```

#### changing the width 

```python
plt.bar(classes,classes1_student,width=0.2)
```

#### align the chart

```python
plt.bar(classes,classes1_student,width=0.2,align='center')
```

#### changing the color

```python
plt.bar(classes,classes1_student,width=0.2,align='edge',color='m')
```

#### changing the color of edges

```python
plt.bar(classes,classes1_student,width=0.2,align='edge',color='m',edgecolor='b')
```

#### changing the line width of the graph

```python
plt.bar(classes,classes1_student,width=0.2,align='edge',color='m',edgecolor='b',linewidth=5)
```

#### changing the alpha, means opacity

```python
plt.bar(classes,classes1_student,width=0.2,align='edge',color='m',edgecolor='b',linewidth=5,alpha=0.9)
```

#### changing the line style

```python
plt.bar(classes,classes1_student,width=0.2,align='edge',color='m',edgecolor='b',linewidth=2,alpha=0.9,linestyle="--")
```

#### giving label

```python
plt.bar(classes,classes1_student,width=0.2,align='edge',
        color='m',edgecolor='b',linewidth=2,alpha=0.9,
        linestyle="--",label='class 1 student')
```

##### to hide the graph

```python
plt.bar(classes,classes1_student,width=0.2,align='edge',
        color='m',edgecolor='b',linewidth=2,alpha=0.9,
        linestyle="--",label='class 1 student',
       visible=False)
```
