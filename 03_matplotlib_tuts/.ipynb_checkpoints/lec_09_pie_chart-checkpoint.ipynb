```python
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]
```

```python
#### classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students

explode=[0.1,0,0,0.2,0]
colors=["c","r","b","y","m"]
textprops={'fontsize':25}

plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
       shadow=True,radius=2.4,startangle=90,
       textprops=textprops)

plt.show()
```

#### pctdistance ->  %distance -> aage pihce krne ke liye ->float me
#### wedgeprops ->  widget property ->line ki size/width km jyada, color change krna hai to use krenge, dict me value dete hai
#### center->  x and y axis pe kidhar plot krna hai m tuple me value
#### frame-> frame True
#### rotatelabels=True

```python
wedgeprops={'linewidth':4,'width':1,'edgecolor':'k'}
plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
       shadow=True,radius=2.4,startangle=90,
       wedgeprops=wedgeprops,frame=True,rotatelabels=True,center=(2,3))

plt.show()
```

### to fix the above graph we have to change the values and learn from output

```python
explode=[0.1,0,0,0.2,0]
colors=["c","r","b","y","m"]
textprops={'fontsize':25}

plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
       shadow=True,radius=2.4,startangle=90,
       textprops=textprops)
plt.legend()
plt.show()
```

#### plotting a unique pie chart like circle

```python
import numpy as np
plt.figure(figsize=(6,5))

colors=['r','w','r','w','r','w','r','w','r','w','r','w',
        'r','w','r','w','r','w','r','w']
labels=np.ones(20)

plt.pie([1],colors="k",radius=2.05)
plt.pie(labels,colors=colors,radius=2.0)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(16,9))

colors=['r','w','r','w','r','w','r','w','r','w','r','w',
        'r','w','r','w','r','w','r','w']
labels=np.ones(20)


plt.pie([1],colors="k",radius=2.05) #outer circle
plt.pie(labels,colors=colors,radius=2.0) #this will create white and red inside the first circle with given label as r/w


# this will draw circle inside the main cicle with their radius
plt.pie([1],colors="g",radius=1.8)
plt.pie([1],colors="y",radius=1.6)
plt.pie([1],colors="c",radius=1.4)
plt.pie([1],colors="b",radius=1.2)
plt.pie([1],colors="m",radius=1.0)
plt.pie([1],colors="r",radius=0.8)
plt.pie([1],colors="y",radius=0.6)
plt.pie([1],colors="b",radius=0.4)

# this will again outer circle with black color
plt.pie([1],colors="b",radius=0.31) 
plt.pie(labels,colors=colors,radius=0.3)  #this will be same as main circle as outer but having different radius

# this is the innermost cirle
plt.pie([1],colors="w",radius=0.2)
plt.pie([1],colors="k",radius=0.1)
plt.show()




```
