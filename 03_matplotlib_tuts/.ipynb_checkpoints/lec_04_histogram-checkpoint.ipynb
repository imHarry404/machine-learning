```python
import matplotlib.pyplot as plt
from matplotlib import style 
import numpy as np
import random
```

```python
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(15,40,(100))
```

```python
print(ml_students_age)
print(py_students_age)
```

###  plotting histogram using these data

```python
plt.hist(ml_students_age)
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)
plt.show()
```

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.hist(ml_students_age,bins)
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)
plt.show()
```

### setting width that will make space between

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.hist(ml_students_age,bins,rwidth=0.8) #80% to show and 20% white space
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)
plt.show()
```

###  hist type

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.hist(ml_students_age,bins,rwidth=0.8,histtype='step')
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)
plt.show()
```

### align mid/left/right

###  orientation-> vertical/horizontal

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.hist(ml_students_age,bins,rwidth=0.8,orientation='horizontal')
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)
plt.show()
```

### changing the color

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.hist(ml_students_age,bins,rwidth=0.8,
         color='m')
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)
plt.show()
```

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.hist(ml_students_age,bins,rwidth=0.8,
         color='y')
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)
plt.show()
```

### label

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.hist(ml_students_age,bins,rwidth=0.8,
         color='m',label='ml-students')
plt.legend()#this will make the label to show
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)

plt.show()
```

### plotting 2 data in one histogram
##### plotting ml students and python students in the same histrigram

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.hist(ml_students_age,bins,rwidth=0.8,
         color='m',label='ml-student')

plt.hist(py_students_age,bins,rwidth=0.8,
         color='r',label='py-student')
plt.legend()
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)
plt.show()
```

### plotting in singe data

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,
         color=['m','b'],label=['ml-student','py-student'])

# plt.hist(py_students_age,bins,rwidth=0.8,color='r',
# label='py-student')


plt.legend()
plt.title('ml students age historgram',fontsize=20)
plt.xlabel('students age category',fontsize=15)
plt.ylabel('no of students age',fontsize=15)
plt.show()
```

#### increasing the size of the histogram

```python
# bins decide the range of graph to show

bins=[15,20,25,30,35,40,45,50]

plt.figure(figsize=(15,9))

plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,
         color=['m','b'],label=['ml-student','py-student'])

# plt.hist(py_students_age,bins,rwidth=0.8,color='r',
# label='py-student')


plt.legend()
plt.title('ml students age historgram',fontsize=30)
plt.xlabel('students age category',fontsize=25)
plt.ylabel('no of students age',fontsize=25)
plt.show()
```

### chjanging the background color using style method

```python
# bins decide the range of graph to show

style.use("ggplot") #ye sbse upr rhega

bins=[15,20,25,30,35,40,45,50]

plt.figure(figsize=(16,9))

plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,
         color=['m','b'],label=['ml-student','py-student'])

# plt.hist(py_students_age,bins,rwidth=0.8,color='r',
# label='py-student')


plt.legend()
plt.title('ml students age historgram',fontsize=30)
plt.xlabel('students age category',fontsize=25)
plt.ylabel('no of students age',fontsize=25)
plt.show()
```
