### colors

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students

explode=[0.03,0,0,0.2,0]
colors=["c","r","b","y","m"]

plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors)

plt.show()
```

#### autopercentage
ye data to percentage value ke sath print krega

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students

explode=[0.03,0,0,0.2,0]
colors=["c","r","b","y","m"]

plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%")

plt.show()
```

#### shadow

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students

explode=[0.1,0,0,0.2,0]
colors=["c","r","b","y","m"]

plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
       shadow=True)

plt.show()
```

#### radius
size bda krne me help aata hai

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students

explode=[0.1,0,0,0.2,0]
colors=["c","r","b","y","m"]

plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
       shadow=True,radius=2.4)

plt.show()
```

#### start angle

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students

explode=[0.1,0,0,0.2,0]
colors=["c","r","b","y","m"]

plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
       shadow=True,radius=2.4,startangle=90)

plt.show()
```

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students

explode=[0.1,0,0,0.2,0]
colors=["c","r","b","y","m"]

plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
       shadow=True,radius=2.4,startangle=270)

plt.show()
```

#### text property

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students

explode=[0.1,0,0,0.2,0]
colors=["c","r","b","y","m"]
textprops={'fontsize':25}

plt.pie(classes1_student,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
       shadow=True,radius=2.4,startangle=90,
       textprops=textprops)

plt.show()
```
