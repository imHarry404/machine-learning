## scatter plot

```python
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import numpy as np
```

```python
df_google_play_store_apps=pd.read_csv('lec_07_ka_csv_file.csv')
df_google_play_store_apps.shape
```

#### taking few data from that

```python
df_google_play_store_apps=pd.read_csv('lec_07_ka_csv_file.csv'
                                     ,nrows=100)
df_google_play_store_apps.shape
```

```python
df_google_play_store_apps
```

##### taking int columns from the data set and plotting the scatter graph

```python
x=df_google_play_store_apps['Rating']
y=df_google_play_store_apps['Reviews']
```

```python
plt.scatter(x,y)
```

#### setting the labels

```python
plt.scatter(x,y)
plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews',fontsize=20)
```

#### modifying the graph

```python
plt.figure(figsize=(16,9)) #ye sbse phle rhega
plt.scatter(x,y,color='r')

plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews',fontsize=20)
```

```python
plt.figure(figsize=(16,9)) #ye sbse phle rhega
plt.scatter(x,y,color='r',marker='*')

plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews',fontsize=20)
```

#### marker size big

```python
plt.figure(figsize=(16,9)) #ye sbse phle rhega
plt.scatter(x,y,color='r',marker='*',s=500)

plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews',fontsize=20)
```

```python
plt.figure(figsize=(16,9)) #ye sbse phle rhega
plt.scatter(x,y,color='r',marker='>',s=500)

plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews',fontsize=20)
```

```python
plt.figure(figsize=(16,9)) #ye sbse phle rhega
plt.scatter(x,y,c='r',marker='^',s=500)

plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews',fontsize=20)
```

#### alpha=-> transparancy

```python
plt.figure(figsize=(16,9)) #ye sbse phle rhega
plt.scatter(x,y,c='r',marker='>',s=500,alpha=0.4)

plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews',fontsize=20)
```

#### line width

```python
plt.figure(figsize=(16,9)) #ye sbse phle rhega
plt.scatter(x,y,c='r',marker='o',s=500,alpha=0.4,linewidths=10)

plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews',fontsize=20)
```

#### hcolor

```python
plt.figure(figsize=(16,9)) #ye sbse phle rhega
plt.scatter(x,y,c='r',marker='o',s=500,alpha=0.6,
            linewidths=10,edgecolor='y')

plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews',fontsize=20)
```

#### plotting multiple data in a single scatter graph

```python
df_google_play_store_apps.head(2)
```

```python
plt.figure(figsize=(16,9)) #ye sbse phle rhega

# 1st
plt.scatter(x,y,c='y',marker='*',s=500,linewidths=10,edgecolor='m',alpha=0.5)

#2nd
plt.scatter(x,df_google_play_store_apps['Installs'],c='r',marker='o',s=500,linewidths=10,edgecolor='m',alpha=0.5)

plt.title('Google play store apps scatter plot',fontsize=15)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('Reviews and Installs',fontsize=20)
```
