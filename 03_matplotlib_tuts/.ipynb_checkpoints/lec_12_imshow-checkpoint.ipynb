#### imshow
it is used to show color and make very unique images efect

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### reading the image

##### mpimg.imread(fname, format=None)

```python
mpimg.imread('circle.png')
```

#### plotting images using that array using plt.imshow()

### plt.imshow(
    X,
    cmap=None,
    norm=None,
    aspect=None,
    interpolation=None,
    alpha=None,
    vmin=None,
    vmax=None,
    origin=None,

```python
plt.imshow(img)
plt.show()
```

#### hidding the axis

```python
plt.axis('off')
plt.imshow(img)
plt.show()
```

#### increasing the size

```python
plt.figure(figsize=(16,9))
plt.axis('off')
plt.imshow(img)
plt.show()
```

##### showing the colorbar of the pic

```python
plt.figure(figsize=(16,9))
plt.axis('off')
plt.imshow(img)
plt.colorbar() #no argument required
plt.show()
```

#### using cmap we can give color
cmap having n no of parameter
it only works with 2d , 

```python
plt.figure(figsize=(16,9))
plt.axis('off')
plt.imshow(img,cmap="Accent")
plt.colorbar() #no argument required
plt.show()
```

####  reading another file and apply the effect

```python
img=mpimg.imread('pie_chart.png')
single_channel=img[:,:,1] #here taking data f
plt.figure(figsize=(16,9))
plt.axis('off')
plt.imshow(single_channel,cmap="hot")
plt.colorbar() #no argument required
plt.show()
```

```python
img=mpimg.imread('pie_chart.png')
single_channel=img[:,:,2] #here taking data f
plt.figure(figsize=(16,9))
plt.axis('off')
plt.imshow(single_channel,cmap="hot")
plt.colorbar() #no argument required
plt.show()
```
