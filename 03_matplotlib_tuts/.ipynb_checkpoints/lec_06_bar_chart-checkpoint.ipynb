```python
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib import style
```

```python
classes=['Python','R','AI','ML','DS']
classes1_student=[30,10,20,25,10]#out of 100 students
classes2_student=[40,5,20,20,10]
classes3_student=[35,5,30,15,15]
```

#### increasing the size of the bar graph

```python
plt.bar(classes,classes1_student,width=0.2,align='edge',
        color='m',edgecolor='b',linewidth=2,alpha=0.9,
        linestyle="--",label='class 1 student')
plt.legend()
```

```python
plt.bar(classes,classes1_student,width=0.6,align='edge',
        color='k',edgecolor='b',linewidth=2,alpha=0.9,
        linestyle="--",label='class 1 student')
plt.legend()
```

#### setting background color usingh style.use

```python
style.use("ggplot")
plt.bar(classes,classes1_student,width=0.6,align='edge',
        color='m',edgecolor='b',linewidth=2,alpha=0.9,
        linestyle="--",label='class 1 student')
plt.legend()
```

```python
style.use("ggplot")

plt.figure(figsize=(16,9)) #ratio

plt.bar(classes,classes1_student,width=0.6,align='edge',
        color='m',edgecolor='b',linewidth=2,alpha=0.9,
        linestyle="--",label='class 1 student')
plt.legend()
```

#### making it horizontal

```python
style.use("ggplot")

plt.figure(figsize=(16,9)) #ratio
# width can't be given in barh
plt.barh(classes,classes1_student,align='edge',
        color='m',edgecolor='b',linewidth=2,alpha=0.9,
        linestyle="--",label='class 1 student')
plt.legend()
```

##### setting title

```python
style.use("ggplot")

plt.figure(figsize=(16,9)) #ratio
# width can't be given in barh
plt.barh(classes,classes1_student,align='edge',
        color='m',edgecolor='b',linewidth=2,alpha=0.9,
        linestyle="--",label='class 1 student')

plt.title('Bar chart of Classes',fontsize=25)
plt.xlabel('subject name',fontsize=25)
plt.ylabel("no of students",fontsize=25)
plt.legend()
```

####  plotting graph with multiple data

```python
style.use("ggplot")

plt.figure(figsize=(16,9)) #ratio

# class1
plt.bar(classes,classes1_student,align='edge',width=0.2,
        color='m',linestyle="--",label='class 1 student')

# class2
plt.bar(classes,classes2_student,align='edge',width=0.2,
        color='b',linestyle="--",label='class 2 student')


plt.title('Bar chart of Classes',fontsize=25)
plt.xlabel('subject name',fontsize=25)
plt.ylabel("no of students",fontsize=25)
plt.legend()
```

#### to draw seperate graph not to override

```python
style.use("ggplot")

# plt.figure(figsize=(16,9)) #ratio


classes_index=np.arange(len(classes))
width=0.2
# class1
plt.bar(classes_index,classes1_student,width,
        color='m',label='class 1 student')

# class2 (+width means phle class1 aayega then utna width chhor ke next aayega)
plt.bar(classes_index+width,classes2_student,width,
        color='b',label='class 2 student')

plt.title('Bar chart of Classes',fontsize=25)
plt.xlabel('subject name',fontsize=25)
plt.ylabel("no of students",fontsize=25)
plt.legend()
```

#### xticks will allow to show the data 

```python
style.use("ggplot")

# plt.figure(figsize=(16,9)) #ratio


classes_index=np.arange(len(classes))
width=0.2
# class1
plt.bar(classes_index,classes1_student,width,
        color='m',label='class 1 student')

# class2 (+width means phle class1 aayega then utna width chhor ke next aayega)
plt.bar(classes_index+width,classes2_student,width,
        color='b',label='class 2 student')

plt.xticks(classes_index+width,classes) #this will allow to show the data of garph

plt.title('Bar chart of Classes',fontsize=25)
plt.xlabel('subject name',fontsize=25)
plt.ylabel("no of students",fontsize=25)
plt.legend()
```

#### rotating the labels


```python
style.use("ggplot")

# plt.figure(figsize=(16,9)) #ratio


classes_index=np.arange(len(classes))
width=0.2
# class1
plt.bar(classes_index,classes1_student,width,
        color='m',label='class 1 student')

# class2 (+width means phle class1 aayega then utna width chhor ke next aayega)
plt.bar(classes_index+width,classes2_student,width,
        color='b',label='class 2 student')

plt.xticks(classes_index+width,classes,rotation=25) #this will allow to show the data of garph

plt.title('Bar chart of Classes',fontsize=25)
plt.xlabel('subject name',fontsize=25)
plt.ylabel("no of students",fontsize=25)
plt.legend()
```

####  plotting graph with  3 data

```python
style.use("ggplot")

# plt.figure(figsize=(16,9)) #ratio


classes_index=np.arange(len(classes))
width=0.2
# class1
plt.bar(classes_index,classes1_student,width,
        color='m',label='class 1 student')

# class2 (+width means phle class1 aayega then utna width chhor ke next aayega)
plt.bar(classes_index+width,classes2_student,width,
        color='b',label='class 2 student')


# class3 (2 times width means phle 2 wala ke bad)
plt.bar(classes_index+width+width,classes3_student,width,
        color='y',label='class 3 student')


plt.xticks(classes_index+width,classes,rotation=25) #this will allow to show the data of garph

plt.title('Bar chart of Classes',fontsize=25)
plt.xlabel('subject name',fontsize=25)
plt.ylabel("no of students",fontsize=25)
plt.legend()
```

####  drawing same graph horizontally
we have to change
xticks to yticks
x_label to y_label



```python
style.use("ggplot")

# plt.figure(figsize=(16,9)) #ratio


classes_index=np.arange(len(classes))
width=0.2
# class1
plt.barh(classes_index,classes1_student,width,
        color='m',label='class 1 student')

# class2 (+width means phle class1 aayega then utna width chhor ke next aayega)
plt.barh(classes_index+width,classes2_student,width,
        color='b',label='class 2 student')


# class3 (2 times width means phle 2 wala ke bad)
plt.barh(classes_index+width+width,classes3_student,width,
        color='y',label='class 3 student')


plt.yticks(classes_index+width,classes,rotation=25) #this will allow to show the data of garph

plt.title('Bar chart of Classes',fontsize=25)
plt.ylabel('subject name',fontsize=25)
plt.xlabel("no of students",fontsize=25)
plt.legend()
```
