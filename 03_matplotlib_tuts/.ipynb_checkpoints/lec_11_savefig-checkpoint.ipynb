### how to save plots(pie,bar)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import style
```

#### savefig
  savefig(fname, dpi=None, facecolor='w', edgecolor='w',
          orientation='portrait', papertype=None, format=None,
          transparent=False, bbox_inches=None, pad_inches=0.1,
          frameon=None, metadata=None)

```python
plt.pie([10,20,30])
plt.savefig('pie_chart') #default is png
plt.show()
```

#### using dpi to handle resolution

```python
plt.pie([10,20,30])
plt.savefig('pie_chart',dpi=100) #default is png
plt.show()
```

#### quality (1-100)

```python
plt.pie([10,20,30])
plt.savefig('pie_chart',dpi=100,quality=99) #default is png
plt.show()
```

##### facecolor='color' -> to set background color

```python
plt.pie([10,20,30])
plt.savefig('pie_chart',dpi=100,quality=99,facecolor="m") #default is png
plt.show()
```

```python
import numpy as np

colors=['r','w','r','w','r','w','r','w','r','w','r','w',
        'r','w','r','w','r','w','r','w']
labels=np.ones(20)

plt.pie([1],colors="k",radius=2.05)
plt.pie(labels,colors=colors,radius=2.0)
plt.legend("brw")
plt.savefig('circle')
plt.show()
```
