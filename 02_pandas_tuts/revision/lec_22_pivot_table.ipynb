```python
import pandas as pd
```

#### pandas.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False, sort=True)[source]
Create a spreadsheet-style pivot table as a DataFrame.

The levels in the pivot table will be stored in MultiIndex objects (hierarchical indexes) on the index and columns of the result DataFrame.

```python
df=pd.read_csv('lec_22_ka_csv_file.csv')
df
```

#### df.pivot_table(
    values=None,
    index=None,
    columns=None,
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
    sort=True,
)

```python
df.pivot_table(index='genre')
```

```python
df.pivot_table(index='genre',columns='year') #missing values will be NaN
```

```python
df.pivot_table(index='genre',columns='year',aggfunc='mean')
```

```python
df.pivot_table(index='genre',columns='year',aggfunc='max')
```

```python
df.pivot_table(index='genre',columns='year',aggfunc='min')
```

```python
df.pivot_table(index='genre',columns='year',aggfunc='count')
```

### filling NaN values using fill_value

```python
df.pivot_table(index='genre',columns='year',aggfunc='count',fill_value='harry')#all NaN value will be filled by harry
```

####  if our data set has one column completly NaN it will be removed

```python
df.pivot_table(index='genre',columns='year',fill_value='harry',margins=True)#margins -> it adds total of rows and column
```

####  using agg func with margins

```python
df.pivot_table(index='genre',columns='year',aggfunc='count',margins=True)
```

```python
df.pivot_table(index='genre',columns='year',aggfunc='sum',margins=True)
```

```python
df.pivot_table(index='genre',columns='year',aggfunc='mean',margins=True)
```

```python
df.pivot_table(index='genre',columns='year',aggfunc='max',margins=True)
```
