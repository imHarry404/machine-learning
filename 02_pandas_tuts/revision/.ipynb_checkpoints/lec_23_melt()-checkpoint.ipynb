```python
import pandas as pd
```

#### pandas.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None, ignore_index=True)[source]¶
Unpivot a DataFrame from wide to long format, optionally leaving identifiers set.

This function is useful to massage a DataFrame into a format where one or more columns are identifier variables (id_vars), while all other columns, considered measured variables (value_vars), are “unpivoted” to the row axis, leaving just two non-identifier columns, ‘variable’ and ‘value’.

```python
df=pd.read_csv('lec_23_ka_csv_file.csv')
df
```

### pd.melt(
    frame: 'DataFrame',
    id_vars=None,
    value_vars=None,
    var_name=None,
    value_name='value',
    col_level=None,
    ignore_index: 'bool' = True,
)

```python
pd.melt(df)
```

```python
pd.melt(df,id_vars='year')
```

```python
pd.melt(df,id_vars=['year','genre'])
```

```python
pd.melt(df,id_vars=['year'],value_vars=['genre']) #year ko genre ke hisab se dekhna
```

```python
pd.melt(df,id_vars=['year'],value_vars=['rating'])
```

```python
pd.melt(df,id_vars=['year'],value_vars=['rating'],var_name='harry') #variable column ka name change kr dega
```

```python
pd.melt(df,id_vars=['year'],value_vars=['rating'],var_name='category') #variable column ka name change kr dega
```

```python
pd.melt(df,id_vars=['year'],value_vars=['rating'],var_name='changed_variable',value_name='changed_value') #variable column ka name change kr dega
```
