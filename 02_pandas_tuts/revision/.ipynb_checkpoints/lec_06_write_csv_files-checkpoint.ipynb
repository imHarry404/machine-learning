####  setting manually header

```python
df=pd.read_csv('lec_06_ka_csv_file.csv',header=1) #1 means first column
df
```

### if our data has first empty line then it will name it as unnamed0/1...so header is used to give custom heading

##### if our data has no heading, then first data will be heading
in this case we will set header=none, then pandas will automatically set the header as 0,1...

##### we can give manually heading using 
header=none,prefix=['column'] it will name it as column0,1.,...

```python
df=pd.read_csv('lec_06_ka_csv_file.csv',header=None) #1 means first column
df
```

```python
df=pd.read_csv('lec_06_ka_csv_file.csv',header=None,prefix='harry')
df
```

###### to identify the column uniquely
we can use names paramter
names=[names for the columns as per no of column in the file as string]

```python
df=pd.read_csv('lec_06_ka_csv_file.csv',names=['a','b','c','d','e','f'])
df
```

```python
df=pd.read_csv('lec_06_ka_csv_file.csv',names=['id','name','single','age','avg','total'])
df
```
