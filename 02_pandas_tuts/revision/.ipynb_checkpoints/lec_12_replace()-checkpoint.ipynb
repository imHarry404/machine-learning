## df.replace()

replace() function is used to replace a string, regex, list, dictionary, series, number etc. from a dataframe. This is a very rich function as it has many variations. The most powerful thing about this function is that it can work with Python regex (regular expressions).

```python
import pandas as pd
```

```python
df=pd.read_csv('lec_12_ka_csv_file.csv')
df
```

#### df.replace(
    to_replace=None,
    value=None,
    inplace: 'bool' = False,
    limit=None,
    regex: 'bool' = False,
    method: 'str' = 'pad',
)

```python
df
```

```python
df.replace(to_replace='iit',value='chaned_iit')
```

```python
df.replace('iit','chaned_iit')
```

```python
df.replace(5.0,15)
```

```python
df.replace([])
```

```python
df.replace([1,2,3,4,5],'changed')
```

```python
df.replace([1,2,3,4,5],99)
```

```python
df.replace([1,2,3,4,5],[10,20,30,40,50])
```

```python
df.replace({'companies':'gitm'},'none')
```

```python
df.replace({'companies':'gitm','profit':36},'none')
```

#### regex

```python
df
```

```python
df.replace('[A-Za-z]',0)
```

```python
df.replace('[A-Za-z]',0,regex=True)#it will replace A-Z a-z values into 0
```

```python
df
```

```python
df.replace({'companies':'[A-Za-z]'},0,regex=True) #changing in particular column
```

```python
df
```

```python
df.replace('gitm',method='ffill') #pichle walw value se replace ho jayega
```

```python
df
```

```python
df.replace('gitm',method='bfill') #pichle walw value se agle ho jayega
```

```python
df
```

```python
df.replace('gitm',method='bfill',limit=1) #1 gitm change ho jayega
```

```python
df.replace('gitm',method='bfill',limit=3) #1 gitm change ho jayega
```

```python
df
```

#### inplace =true krne ke bad modified df save ho jata hai


```python
df.replace('gitm',method='bfill',limit=3,inplace=True)
```
