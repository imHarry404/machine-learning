#### DataFrame.append(other, ignore_index=False, verify_integrity=False, sort=False)[source]Â¶
Append rows of other to the end of caller, returning a new object.

Columns in other that are not in the caller are added as new columns.

```python
df1=pd.DataFrame({'a':[1,2,3],
                  'b':[4,5,6]})
df2=pd.DataFrame({'a':[7,8,9],
                  'b':[10,11,12]})
display(df1,df2)
```

```python
df1.append(df2) #here the index are not in proper we need to false the index
```

```python
df1.append(df2,ignore_index=True)
```

```python
df2.append(df1,ignore_index=True)
```

### if data frame has different column name

```python
df1=pd.DataFrame({'a':[1,2,3],
                  'b':[4,5,6]})
df2=pd.DataFrame({'c':[7,8,9],
                  'd':[10,11,12]})
display(df1,df2)
```
