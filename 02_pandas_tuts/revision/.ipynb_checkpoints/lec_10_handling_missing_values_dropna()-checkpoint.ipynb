```python
import pandas as pd
```

```python
df=pd.read_csv('lec_10_ka_csv_file.csv')
df
# here we have some missing values
```

#### dropna()-> if any row has nan value it will remove that  whole row

```python
df.dropna()
```

#### df.dropna(
    axis: 'Axis' = 0,
    how: 'str' = 'any',
    thresh=None,
    subset=None,
    inplace: 'bool' = False,
)

```python
df.dropna(axis=1)#means it will drop those rows 
```

```python
df.dropna(how='any')#any is default, it will remove those rows which has nan values
```

```python
df.dropna(axis=1,how='any')
```

```python
df.dropna(how='all') #if data set has all nan values then it will drop but if few are nan then it will do nothing
```

```python
df=pd.read_csv('lec_10_ka_csv_file.csv')
df
```

```python
df.dropna(how='all')  #it will remove the last row
```

###  thresh-> int (None)
#### to know how many non values are in an particular row

```python
df=pd.read_csv('lec_10_ka_csv_file.csv')
df
```

```python
df.dropna(thresh=1) #last will be avoid becuase all are nan value
```

```python
df.dropna(thresh=4) #here it will check that which row has atleast 4 nan value
```

```python
df
```

###### subset
jb kisi specific column ke nan value wale cell ko drop krna ho

```python
df.dropna(subset=['id'])
```

```python
df.dropna(subset=['name'])# name me se NaN vala data ht jayega
```

```python
df.dropna(subset=['profit'])
```
