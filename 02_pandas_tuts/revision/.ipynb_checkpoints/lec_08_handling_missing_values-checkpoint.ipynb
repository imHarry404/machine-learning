### handling missing values in python
na_values   
keep_default_na   
na_filter

```python
import pandas as pd
```

```python
df=pd.read_csv('lec_08_ka_csv_file.csv')
df
# it will print Nan at black cells means not a number
```

```python
df=pd.read_csv('lec_08_ka_csv_file.csv')
df
# it will print Nan at black cells means not a number
```

##### to make any string as NaN we use na_values

```python
df=pd.read_csv('lec_08_ka_csv_file.csv')
df
# it will print Nan at black cells means not a number
```

#### conmsidering not available as NaN

```python
df=pd.read_csv('lec_08_ka_csv_file.csv',na_values='not available')# it will print Nan at black cells means not a number
df
```

##### to make NaN more than one word we use list

```python
df=pd.read_csv('lec_08_ka_csv_file.csv')
df
```

```python
df=pd.read_csv('lec_08_ka_csv_file.csv',na_values=['no values','not available','nothing'])
df
```

####  to consider NaN of any specific column
we will give them in dictionary

```python
df=pd.read_csv('lec_08_ka_csv_file.csv',na_values={'total':'#na'})
df
# here in total column the #na value will be consiter as NaN
```

#### keep_default_na

```python
df=pd.read_csv('lec_08_ka_csv_file.csv')
df
```

```python
df=pd.read_csv('lec_08_ka_csv_file.csv',keep_default_na=True)
df
```

```python
df=pd.read_csv('lec_08_ka_csv_file.csv',keep_default_na=False)
df
```

#### filter
this is used to find the missing values

```python
df=pd.read_csv('lec_08_ka_csv_file.csv',na_filter=False)
df
```
