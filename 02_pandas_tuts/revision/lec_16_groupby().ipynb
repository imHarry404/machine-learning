####  panads groupby()
it is used to split the data in group as per some criteria


```python
import pandas as pd
```

```python
df=pd.read_csv("lec_16_ka_csv_file.csv")
df
```

##### df.groupby(
    by=None,
    axis: 'Axis' = 0,
    level: 'Level | None' = None,
    as_index: 'bool' = True,
    sort: 'bool' = True,
    group_keys: 'bool' = True,
    squeeze: 'bool | lib.NoDefault' = <no_default>,
    observed: 'bool' = False,
    dropna: 'bool' = True,
)

```python
gr1=df.groupby(by='section')
gr1
```

```python
gr1.groups
```

```python
gr2=df.groupby(by=['section','class'])
gr2
```

```python
gr2.groups
```

```python
for data in gr2.groups:
    print(data)
```

```python
df
```

```python
for section, df_1 in gr2:
    print(section)
    print(df_1)
```
