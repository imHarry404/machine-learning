```python
import pandas as pd
```

```python
df1=pd.DataFrame({'id':[1,2,3,4],'class':[6,7,8,9],'name':['a','b','c','d']})
df1
```

```python
df2=pd.DataFrame({'id':[1,2],'class':[10,11],'name':['e','f']})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

####  to print the common values from both data frame

```python
pd.concat([df1,df2],axis=1,join='outer') #outer means union means all
```

```python
pd.concat([df1,df2],axis=1,join='inner') #means intersection means common from both
```

#### comining the data frame using keys =[]

```python
pd.concat([df1, df2],keys=['df1,df2'])
```

```python
pd.concat([df1, df2],keys=['first df,second df'])
```

```python
pd.concat([df1, df2],axis=1,keys=['first df,second df'])
```
