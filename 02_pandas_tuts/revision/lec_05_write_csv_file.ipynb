#### to read the limited no of rows  using nrow=value

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',nrows=1)
df
```

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',nrows=5)
df
```

####  to see the specific given columns using usecols[index]

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',usecols=[0])
df
```

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',usecols=[0,1,5,7])
df
```

#### we cant print selected columns from the file using usecols=[], we just have to give the index of those columns

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',usecols=[4,8])
df
```

######  we can make rows as heading using skiprows=1, means after skip row the next row will be heading

```python
df=pd.read_csv('lec_05_ka_csv_file.csv')
df
```

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',skiprows=1)
df
```

####  to skip the specific row

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',skiprows=[2])
df
```

#### we can skip multiple rows, by giving their index

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',skiprows=[1,6,8])
df
```

#### to change the default index
we can customize it using index_col
means making column as row/index

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',index_col='p21')
df
```

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',index_col='Number')
df
```

#### by giving index we can make it as column

```python
df=pd.read_csv('lec_05_ka_csv_file.csv',index_col=5)
df
```
