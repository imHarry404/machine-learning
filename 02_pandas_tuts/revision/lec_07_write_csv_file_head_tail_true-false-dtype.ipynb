##### if 'yes' is written in our file then true_value and if we want to change it into string , it will write True at yes so that we can find that column specifically

```python
df=pd.read_csv('lec_07_ka_csv_file.csv')
df
```

```python
# it will change all yes into true
df=pd.read_csv('lec_07_ka_csv_file.csv',true_values=['yes'])
df
```

##### false_values will not work without true_values

```python
# it will change all yes into true
df=pd.read_csv('lec_07_ka_csv_file.csv',true_values=['yes'],false_values=['no'])
df
```
