# pandas interpolate()

```python
import pandas as pd
```

interpolate fuction is used to fill missing values/ NaN in df or series,
syntax pd.interpolate(), series.interpolate()

```python
df=pd.read_csv('lec_13.csv')
```

```python
df
```

this method guesses the missing values and fill only numeric values, it cann't guess string values

```python
df.interpolate()
# df.interpolate(method=linear) same
```

df.interpolate(method=time)
this method will work only when the data type will be in date_time format

```python
type(df.Date[0])
```

this column is in str we have to read it and parse it into datetime so that we can use time method of interpolate

```python
df=pd.read_csv('lec_13.csv',parse_dates=['Date'])
```

```python
df
```

now the date column will be pandas._libs.tslibs.timestamps.Timestamp so that we can use time method od interpolate 


```python
type(df.Date[0])
```

```python
df
```

changing the default column with our date column

```python
df=pd.read_csv('lec_13.csv',parse_dates=['Date'],index_col='Date')
```

```python
df
```

```python
df.interpolate(method='time')
# it will fill the missing values according to dateTime, with nearest possible value by guessing
```

```python
df.interpolate(method='index')
# it will fill the values according to index
# it will be silimar to date becuase our index col is date
```

```python
df=pd.read_csv('lec_13.csv',index_col='class')
# before this we must delete the string columns
```

```python
df
```

```python
df.interpolate(method='index')
# it will fill the missing values as per class
```
