```python
import pandas as pd
```

reading a csv file

```python
df=pd.read_csv('lec_05.csv')
df
```

type of read file and reading specific row or columns

```python
type(df)
```

reading specifc no of rows or specific rows using index

```python
df=pd.read_csv('lec_05.csv',nrows=2) #only two rows
df
```

reading specific column or columns using index

```python
original_file=pd.read_csv('lec_05.csv')
original_file
```

```python
df=pd.read_csv('lec_05.csv',usecols=[0]) #only 0th index
df
```

```python
df=pd.read_csv('lec_05.csv',usecols=[0,2]) #only 0th and 2nd index
df
```

to sikp the specific or given rows while reading

```python
original_file
```

```python
df=pd.read_csv('lec_05.csv',skiprows=2) #it will skip 2 rows from top
df
```

skipping rows using their index value

```python
df=pd.read_csv('lec_05.csv',skiprows=[1,3,6]) #it will skip 1,3 and 6th rows from top
df
```

changing the index name manually while reading

```python
original_file
```

```python
df=pd.read_csv('lec_05.csv',index_col='Year') #it will make year as its index, default 0,1...will be remove}
df
```
