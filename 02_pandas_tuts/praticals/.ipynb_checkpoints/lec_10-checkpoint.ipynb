```python
df=pd.read_csv('lec_10.csv')
df
```

df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)\n
default axis =0\n
drop  na is used to remove the nan/missing value\n
how='any\all' all means if the whole dataset will be removed or a whole row/column\n
thresh(int)-:> how many minimum not null values we want
subset=['colname'] particular column without nan will be printed
inplace=True\Fa;lse , in case of true it will update the same data set and all nan will be remove
