```python
import pandas as pd
```

```python
df=pd.read_csv('lec_08.csv')
df
```

suppose our file have some missing/blank cells then it will print NaN at those places

```python
df
```

suppose our file has some additonal information that tells that data not available, pandas will print those as same, to make them as NaN we have to use na_values='string' ; all matching string will be considered as NaN

```python
df=pd.read_csv('lec_08.csv')
df
```

```python
df=pd.read_csv('lec_08.csv',na_values='not available') #all not available matching string will be converted into NaN
df
```

to change all types of values into NaN , we will give all those string as list 

```python
df=pd.read_csv('lec_08.csv',na_values=['not available','no value']) #all not available and no value matching string will be converted into NaN
df
```

suppose our file has more type of values which means data not avaible then we will use dictionary and then using col_name we will change the data into NaN

```python
df=pd.read_csv('lec_08.csv')
df
```

```python
df=pd.read_csv('lec_08.csv',na_values={'class':'not available'}) #class clumn ka not available wala NaN ho jayega
df
```

kisi ke v heading capital me ni hona chaiye and space v ni hona chahiye

```python
df=pd.read_csv('lec_08.csv',na_values={'class':'not available','study':'no value','social':'no value','percentege':['not available','no value']}) #given column ka sb NaN ho jayega

df
```

to keep default string , means to print those string which is considred as NaN in  pandas, we will use keep_default_na to print those unsupported strings

#N/A -nan null #NA/NA -nan n/a #na N/A nan -1.#IND NA 1#IND -1.#QNAN NULL 1.QNAN 
these are default considred as NaN by pandas, but we want to print thse values 

```python
df=pd.read_csv('lec_08.csv')
df
```

```python
df=pd.read_csv('lec_08.csv',keep_default_na=False)#default it True
df
```

to filter the missing values, to prase the data in efficient way

```python
df=pd.read_csv('lec_08.csv')
df
```

```python
df=pd.read_csv('lec_08.csv',na_filter=False)#default it True
df
```
