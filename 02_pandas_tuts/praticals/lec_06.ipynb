# how to write csv file in python

header-> this parameter is used when we want to make our heading of dataframe from data

```python
import pandas as pd
```

```python
df=pd.read_csv('lec_05.csv')
df
```

creating customazie heading using row index

```python
df=pd.read_csv('lec_05.csv',header=1)
df
```

if our file has no header , just give header=None, it will make 0,1,2,....default

```python
df=pd.read_csv('lec_05.csv',header=None)
df
```

if we want to give manually heder name we use prefix 
it will make all header by given para

```python
df=pd.read_csv('lec_05.csv',header=None,prefix='header')
df
```

to give desired /customized col name we use name parameter
before using name parameter we must know the no of cols then after we use name parameter
we use list to give the col name

```python
df=pd.read_csv('lec_05.csv')
df
```

```python
df=pd.read_csv('lec_05.csv',names=['first','second','third','forth'])
df
```
