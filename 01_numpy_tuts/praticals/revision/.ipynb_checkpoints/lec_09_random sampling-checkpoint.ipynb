## random sampling using numpy

```python
import numpy as np
import random 
```

#### random.random(int) it will give value in array in 0 to 1

```python
np.random.random(1)
```

```python
np.random.random((3,4))
```

##### np.random.randint(range) -> gives int value in given range 
it will give different value at every execution

```python
np.random.randint(1,4)
```

```python
np.random.randint(1,4,(4,4))
```

```python
np.random.randint(1,4,(4,4,4))
```

#### to use the previous value generated by random
we have to use seed(val)

```python
np.random.seed(10)
np.random.randint(1,4,(4,4))
```

```python
np.random.seed(10)
np.random.randint(1,4,(4,4))
```

#### seed can take 2pow 32-1 values
2**32-1

```python
2**32-1
```
